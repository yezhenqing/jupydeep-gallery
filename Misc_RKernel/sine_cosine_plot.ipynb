# Sine and Cosine Curves

This notebook demonstrates how to plot **sine** and **cosine** curves using `NumPy` for numerical computation and `Matplotlib` for visualization.

The mathematical functions plotted are:

$$\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n+1)!} x^{2n+1}$$

$$\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n}$$

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

print(f"NumPy version: {np.__version__}")
print("Libraries imported successfully.")
```

```python
# Generate x values from 0 to 2π (1000 points for smooth curves)
x = np.linspace(0, 2 * np.pi, 1000)

# Compute sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

print(f"x range: [{x.min():.4f}, {x.max():.4f}]")
print(f"sin(x) range: [{y_sin.min():.4f}, {y_sin.max():.4f}]")
print(f"cos(x) range: [{y_cos.min():.4f}, {y_cos.max():.4f}]")
```

```python
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot sine and cosine curves
ax.plot(x, y_sin, label=r"$\sin(x)$", color="blue", linewidth=2)
ax.plot(x, y_cos, label=r"$\cos(x)$", color="red", linewidth=2, linestyle="--")

# Mark key points where sin and cos intersect
intersections = np.where(np.isclose(y_sin, y_cos, atol=1e-2))[0]
ax.scatter(x[intersections], y_sin[intersections], color="green", s=50, zorder=5, label="Intersections")

# Configure axes and labels
ax.set_title("Sine and Cosine Curves", fontsize=16, fontweight="bold")
ax.set_xlabel("x (radians)", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.2, 1.2)
ax.set_xticks(
    [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi],
    [r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"],
)
ax.axhline(0, color="black", linewidth=0.5)
ax.grid(True, linestyle=":", alpha=0.7)
ax.legend(loc="upper right", fontsize=11)

plt.tight_layout()
plt.show()
```
